```python
def conv(duree):
    '''Fonction permettant de convertir les données traduisant les dates de l'année au format adapté pour le traitement en Python. 
    Cette fonction convertit la variable duree (str) donnant une durée en années et en mois en la variable
    res (float) donnant cette même durée en année
    si la données est indéfini, elle est remplacée par np.inf'''
    duree = str(duree)
    duree = duree.replace('s', '').replace(' ', '')
    annees = 0
    mois = 0
    controle = True
    if 'nan' in duree:
        annees = np.inf
        controle = False
    if ('an' not in duree) and ('moi' not in duree):
        annees = np.inf
    if 'an' in duree and controle:
        l = list(duree)
        nbr =''
        car=''
        while car != 'a':
            nbr+=car
            car = l.pop(0) 
        annees = float(nbr)
    if 'moi' in duree:
        try:    
            l.pop(0)
        except:
            l = list(duree)
        nbr=''
        car=''
        while car!='m':
            nbr+=car
            car = l.pop(0)
        mois = float(nbr)/12
    
    return annees + mois
```

```python
duree = df['durée']
duree2 = []
for ligne in duree :
        duree2.append(conv(ligne))

df['durée'] = duree2
```

```python
#df.drop('durée', axis = 0, inplace = True)
df = df[['index', 'fonction', 'cat', 'grade', 'Echelon', 'durée', 'Indice Brut', 'Indice majoré', 'salaire brut', 'indice majoré calculé']]
df.head()
```

```python
pts = 4.686025 #Valeur du point d'indice au moment où les données ont été recueillies 

IM = df['salaire brut']/pts

df['indice majoré calculé'] = round(IM, 2)

df.head()
```

```python
metiers = df['fonction'].unique()

for metier in metiers :
    grille = df.loc[df['fonction'] == metier]
    grades = grille['grade'].unique()
    for grade in grades :
        grille_grade = grille.loc[grille['grade'] == grade]
        # print(df.loc[(df['fonction'] == metier) & (df['grade'] == grade),'Echelon'])
        df.loc[(df['fonction'] == metier) & (df['grade'] == grade), 'Echelon'] = np.array(range(1,grille_grade['grade'].count()+1))
        
        echelon_max = df.loc[(df['fonction'] == metier) & (df['grade'] == grade), 'Echelon'].max()
        df.loc[(df['fonction'] == metier) & (df['grade'] == grade) & (df['Echelon'] == echelon_max), 'durée'] = np.inf
```

```python
df.to_csv(dossier + 'corr_' + fichier, sep = ',')
```

```python
import pandas as pd

# Mots-clés à utiliser pour la sélection
mots_cles = ["Conseiller d'État",
'Maître des requêtes',
'Rapporteur public',
"Auditeur au Conseil d'État" ,
'Secrétaire général' ,
'Secrétaire administratif des administrations',
'Commissaire du gouvernement',
'Chef de service',
'Juriste',
'Attaché de recherche',
'Chargé de mission',
'Administrateur',
'Conseiller juridique',
'Expert en contentieux administratif',
'Responsable de communication',
'Contrôleur de gestion',
'Magistrat',
'Juge adminsitratif',
"Conseil d'Etat",
'Cours nationale']

# Création d'un masque booléen pour filtrer les lignes contenant les mots-clés
masque = df['fonction'].str.contains('|'.join(mots_cles), case=False)  # Remplacez 'colonne' par le nom de la colonne à filtrer

# Application du masque pour sélectionner les lignes correspondantes
df_selection = df[masque]

# Affichage du DataFrame avec les lignes sélectionnées
print(df_selection)
# Grâce à ce premier code on passe déjà de 11275 à 841 métiers, on peut encore réduire en posant certains mots clés comme exclus

```

```python
mots_cibles = ['Police','CESE','muséologie','maritimes', 'Polytechnique', 'ENA','aviation','maritime','marine','comptes','emploi fonctionnel','Emploi fonctionnel','emplois fonctionnels',"corps en voie d'extinction", 'en extinction','Versailles',"(en voie d'extinction)"]  # Mots-clés spécifiques à rechercher
colonne_cible = 'fonction'  # Colonne dans laquelle rechercher les mots-clés

# Créer un masque pour chaque mot-clé et les combiner avec l'opérateur logique |
mots_cibles_masque = df_selection[colonne_cible].str.contains('|'.join(mots_cibles), case=False)

# Appliquer le masque pour filtrer les lignes
nouveau_dataframe = df_selection[~mots_cibles_masque]

# Afficher le nouveau DataFrame sans les lignes contenant les mots-clés
print(nouveau_dataframe)

```
